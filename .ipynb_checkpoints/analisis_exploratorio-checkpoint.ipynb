## RELACIONES CATEGORICAS CON EL PRECIO DE VENTA

#### CALIDAD DE LAS CASAS VS PRECIO DE VENTA

```python
var="OverallQual"
```

```python
data=pd.concat([df_train["SalePrice"],df_train[var]], axis=1) # a lo largo de eje verticula axis=1
data.head(5)
```

#### En dos variables f, ax se guarda el resultado de un plt.subplots, donde se define la nueva figura donde existira la grafica que se va a crear F y ax, es la figura y los ejes que devuelven la instruccion

```python
f, ax= plt.subplots(figsize=(8,6))# figura con dos ejes que va a medir 8x6, tamaño, es arbitrario, matplotlib para area graph
fig=sns.boxplot(x=var,y="SalePrice",data=data) # La info que tendra el diagrama, para grafica en si se usa seaborn
fig.axis(ymin=0,ymax=800000) # defino valores min y max del grafico
```

#### Se puede ver que la calidad es muy importante para definir el precio de las casas, ya que la relaciòn es creciente, tambien se van abriendo los anchos de las cajas, eso quiere decir a que entre mas calidad hay mayor disparidad entre los precios de las casas, por eso hay casas con una calidad de 9 que valen mas que algunas de calidad de 10

### Confirmar nuestra intuicion con esas variables

#### Matriz de correlacion

```python
corr=df_train.corr() ### matriz de correlacion
```

```python
corr
```

```python
f, ax = plt.subplots(figsize=(12,9))
sns.heatmap(corr, vmax=.8,square = True) #para que sea cuadrado y .8 para que el gradiente no llegue hasta uno que es blanco
```

#### Todo lo que esta claro hace ruido y son las variables que nos llaman la atencion

#### Al mirar la fila de saleprice tiene relacion fuerte con calidad, el area vivible, un poco con totalbsmts y año de construccion, sin embargo, hay otras variables que tambien muestran relacion

#### A continuacion, se seleccionaran 10 variables para ver mejor la relacion:

```python
k=10
cols=corr.nlargest(k,"SalePrice")["SalePrice"].index
cols ## diez variables mas relacionadas
```

```python
cm= np.corrcoef(df_train[cols].values.T) #MATRIZ DE CORRELACION, en arreglos, coeficientes de correlacion, T=transponer
cm
```

```python
sns.set(font_scale=1.25) #caracteristica de la fuente 1.25 "el tamaño de los numeros"
```

```python
hm=sns.heatmap(cm,cbar=True,annot=True,square=True,fmt=".2f", annot_kws={"size":10,},yticklabels=cols.values,xticklabels=cols.values)
plt.show()
```

#### El garaje es una variable que no estabamos considerando y es importante, el problema es que estan tan correlacionadas ".88" que podria indicar un problema de colinealidad, es decir, no puede crecer una sin crecer la otra. A mayor area en un garaje màs carros se pueden meter.
#### Pasa de igual forma con el tamaño del primer piso y el tamaño del sotano, estan relacionados ".82", porque el primer piso esta construido sobre el sotano, por ende, solo se debe incluir una de las dos.
