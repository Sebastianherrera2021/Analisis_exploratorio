## Missing data

#### Â¿QuÃ© tan prevalentes son los datos que faltan?
#### Â¿Los datos que faltan son aleatorios o tienen un patrÃ³n?

#### Estas preguntas son necesarias porque la falta de datos puede implicar una reducciÃ²n del tamaÃ±o de la muestra que puede afectar el analisis.

#### Y nos permite asegurarnos de que el proceso de datos faltantes no esta sesgado y oculte una verdad inconveniente.

```python
#missing data
total = df_train.isnull().sum().sort_values(ascending=False) ## datos faltantes
percent = (df_train.isnull().sum()/df_train.isnull().count()).sort_values(ascending=False) # datos faltantes en %
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent']) ## la union de ambas cols creadas
missing_data.head(20)
```

#### Consideraremos que cuando falta mÃ¡s del 15% de los datos, debemos eliminar la variable correspondiente y pretender que nunca existiÃ³.

#### Esto quiere decir que no intentaremos ningÃºn truco para rellenar los datos que faltan en estos casos. 'PoolQC', 'MiscFeature', 'Alley', etc.) 

#### Â¿perderemos estos datos? No. Aunque ninguna de estas variables parece ser muy importante por lo que son aspectos en los que no pensamos a la hora de comprar y quizas por eso faltan datos.

#### podrÃ­amos decir que variables como 'PoolQC', 'MiscFeature' y 'FireplaceQu' son fuertes candidatas para valores atÃ­picos, por lo que estaremos encantados de eliminarlas.

#### Las variables que tienen la palabra garage y bsmt tienen cada una el mismo numero de datos faltantes: 5% y 2.7 % "numero bajo", cuando se repite puede deberse a que son el mismo conjunto de observaciones, en estos casos tomo la variable mas representativa de cada uno y elimino el resto que tienen el mismo % de missing values.

#### Respecto a 'MasVnrArea' y 'MasVnrType', podemos considerar que estas variables no son necesarias. 

#### Finalmente, tenemos una observaciÃ³n que falta en 'Electrical'. Dado que es solo una observaciÃ³n, eliminaremos esta observaciÃ³n y conservaremos la variable.

#### En resumen, para manejar los datos faltantes, eliminaremos todas las variables con datos faltantes, excepto la variable 'Electrical'. En 'ElÃ©ctrico' simplemente eliminaremos la observaciÃ³n con datos faltantes.

```python
#dealing with missing data
df_train = df_train.drop((missing_data[missing_data['Total'] > 1]).index,1) # elimina variables que tienen mas de un dato faltante
df_train
df_train = df_train.drop(df_train.loc[df_train['Electrical'].isnull()].index) #elimina dato faltante de esa variable
df_train.isnull().sum().max() #just checking that there's no missing data missing... 
df_train
```

## Analisis Univariado

En este contexto, la estandarizaciÃ³n de datos significa convertir valores de datos para que tengan una media de 0 y una desviaciÃ³n estÃ¡ndar de 1.

```python
#standardizing data "del precio de venta"
saleprice_scaled = StandardScaler().fit_transform(df_train['SalePrice'][:,np.newaxis]); ## datos estandarizados
low_range = saleprice_scaled[saleprice_scaled[:,0].argsort()][:10] # valores de rango bajo no estan alejados de 0
high_range= saleprice_scaled[saleprice_scaled[:,0].argsort()][-10:] # Valores de rango alto estan lejos de cero
print('outer range (low) of the distribution:')
print(low_range)
print('\nouter range (high) of the distribution:')
print(high_range)
```

## En la bÃºsqueda de la normalidad

#### - Histograma  Curtosis y asimetrÃ­a.
#### - GrÃ¡fica de probabilidad normal: la distribuciÃ³n de datos debe seguir de cerca la diagonal que representa la distribuciÃ³n normal.

```python
#histogram and normal probability plot
sns.distplot(df_train['SalePrice'], fit=norm); # grafica de distribucion de frecuencia
fig = plt.figure()
res = stats.probplot(df_train['SalePrice'], plot=plt) # segundo grafico de probabilidad
```

### 'SalePrice' no es normal. Muestra 'pico', asimetrÃ­a positiva o sesgo a la derecha y no sigue la lÃ­nea diagonal.
#### Pero no todo estÃ¡ perdido. Una simple transformaciÃ³n de datos puede resolver el problema. Esta es una de las cosas asombrosas que puede aprender en los libros de estadÃ­stica: en caso de asimetrÃ­a positiva, las transformaciones logarÃ­tmicas generalmente funcionan bien.

```python
#applying log transformation
df_train['SalePrice'] = np.log(df_train['SalePrice'])
```

```python
#transformed histogram and normal probability plot
sns.distplot(df_train['SalePrice'], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train['SalePrice'], plot=plt)
```

### Veamos quÃ© estÃ¡ pasando con 'GrLivArea'.

```python
#histogram and normal probability plot
sns.distplot(df_train['GrLivArea'], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train['GrLivArea'], plot=plt)
```

#### parece que hay algo de asimetria

```python
#data transformation
df_train['GrLivArea'] = np.log(df_train['GrLivArea'])
```

```python
#transformed histogram and normal probability plot
sns.distplot(df_train['GrLivArea'], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train['GrLivArea'], plot=plt)
```

### y ahora TotalBsmtSF:

```python
#histogram and normal probability plot
sns.distplot(df_train['TotalBsmtSF'], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train['TotalBsmtSF'], plot=plt)
```

#### Hay un nÃºmero importante de observaciones con valor cero (viviendas sin sÃ³tano). Lo cual es un gran problema porque el valor cero no nos permite hacer transformaciones logarÃ­tmicas. Para aplicar aquÃ­ una transformaciÃ³n logarÃ­tmica, crearemos una variable que pueda tener el efecto de tener o no tener sÃ³tano (variable binaria).

#### Luego, haremos una transformaciÃ³n logarÃ­tmica a todas las observaciones distintas de cero, ignorando aquellas con valor cero. De esta forma podemos transformar los datos, sin perder el efecto de tener o no base. Puede que este enfoque no sea el correcto pero se puede intentar..

```python
#create column for new variable (one is enough because it's a binary categorical feature)
#if area>0 it gets 1, for area==0 it gets 0
df_train['HasBsmt'] = pd.Series(len(df_train['TotalBsmtSF']), index=df_train.index)
df_train['HasBsmt'] = 0  # no tiene sotano
df_train.loc[df_train['TotalBsmtSF']>0,'HasBsmt'] = 1 # tiene sotano
```

```python
#transform data
df_train.loc[df_train['HasBsmt']==1,'TotalBsmtSF'] = np.log(df_train['TotalBsmtSF']) ## la transformacion
```

```python
#histogram and normal probability plot
sns.distplot(df_train[df_train['TotalBsmtSF']>0]['TotalBsmtSF'], fit=norm); # no se estan teniendo en cuenta los 0
fig = plt.figure()
res = stats.probplot(df_train[df_train['TotalBsmtSF']>0]['TotalBsmtSF'], plot=plt)
```

#### El mejor enfoque para probar la homocedasticidad de dos variables mÃ©tricas es grÃ¡ficamente., Las desviaciones de una dispersiÃ³n igual se muestran mediante formas como conos (pequeÃ±a dispersiÃ³n en un lado del grÃ¡fico, gran dispersiÃ³n en el lado opuesto) o diamantes (un gran nÃºmero de puntos en el centro de la distribuciÃ³n).

```python
#scatter plot
plt.scatter(df_train['GrLivArea'], df_train['SalePrice']);
```

#### Las versiones anteriores de este diagrama de dispersiÃ³n (anteriores a las transformaciones de registros) tenÃ­an una forma cÃ³nica. Como puede ver, el grÃ¡fico de dispersiÃ³n actual ya no tiene forma cÃ³nica.Â¡Ese es el poder de la normalidad! Solo asegurando la normalidad en algunas variables, resolvimos el problema de homocedasticidad.
